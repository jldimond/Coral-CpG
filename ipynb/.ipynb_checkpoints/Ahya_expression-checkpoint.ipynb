# *Acropora hyacinthus* gene expression data

This workflow downloads and processes *A. hyacinthus* [gene expression data](http://palumbi.stanford.edu/data/Barshis_etal_33496contigs_deduped_all_scaledounts_Nov2012.txt) for subsequent analysis in R. All samples are included in a single file.

```python
cd ../analyses/Ahya
```

```python
!curl -O http://palumbi.stanford.edu/data/Barshis_etal_33496contigs_deduped_all_scaledounts_Nov2012.txt
```

```python
!head -20 Barshis_etal_33496contigs_deduped_all_scaledounts_Nov2012.txt
```

##The original file format was not amenable to analysis, so it was manually copied in Excel, then pasted into a plain text file in TextWrangler, and saved as Barshis_expression.txt.

```python
!head Barshis_expression.txt
```

```python
#Removing header
!sed '1d' Barshis_expression.txt > Ahya_exp
!head Ahya_exp
```

```python
# Sorting before joining
!sort Ahya_exp > Ahya_exp.sorted
!head Ahya_exp.sorted
```

```python
# Joining with CpG O/E
!join Ahya_exp.sorted ID_CpG.sorted > Ahya_exp_CpG
!head Ahya_exp_CpG
```

```python
!tr ' ' \\t < Ahya_exp_CpG > Ahya_exp_CpG2
!head Ahya_exp_CpG2
```
